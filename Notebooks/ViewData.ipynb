```python
import json
import os
```

```python
fileList = ['../Datasets/' + f for f in os.listdir('../Datasets') if f.endswith('.json')]
```

```python
issueEventList = []
```

```python
for f in fileList:
    fileObj = open(f,)
    data = json.load(fileObj)
    for event in data:
        if("type" in event and event["type"] == "IssuesEvent"):
            if(event["payload"]["action"] == "closed"):
                issueEventList.append(event)
    fileObj.close()
```
